```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline

plt.style.use('default')
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.plot(y,x)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Simple Line Plot")
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=y1*y2
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.plot(x,y3,label='sin(x)*cos(x)')
plt.legend()
plt.xlabel('X Axis')
plt.ylabel("X Axis")
plt.title("Multiple Function Ploted")
plt.grid(True,alpha=0.3)
plt.show()
```

```python
np.random.seed(42)
x=np.random.rand(100)
y=2*x+np.random.rand(100)*0.5
plt.scatter(x,y,alpha=0.5)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Scatter Plot')
plt.grid(True,alpha=0.5)
plt.show()
```

```python
np.random.seed(42)
x=np.random.rand(100)
y=np.random.rand(100)
colors=np.random.rand(100)
sizes=1000*np.random.rand(100)

plt.figure(figsize=(10,6))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.6,cmap='plasma')
plt.colorbar(label='Color Intensity')
plt.xlabel('X Axis')
plt.ylabel('Y label')
plt.title("Scatter Plot with Colors and Sizes")
plt.grid(True,alpha=0.3)
plt.show()
```

```python
catagory=['A','B','C','D','E']
values=[23,45,56,78,32]

plt.figure(figsize=(8,6))
plt.bar(catagory,values,color='steelblue',edgecolor='black',linewidth=1.2)
plt.xlabel("Catagories")
plt.ylabel("Values")
plt.title("Simple Bar Chat")
plt.grid(True,alpha=0.3,axis='y')
plt.show()
```

```python
catagory=['A','B','C','D','E']
values=[23,45,56,78,32]

plt.figure(figsize=(8,6))
plt.barh(catagory,values,color='steelblue',edgecolor='black',linewidth=1.2)
plt.xlabel("Catagories")
plt.ylabel("Values")
plt.title("Simple Bar Chat")
plt.grid(True,alpha=0.3,axis='y')
plt.show()
```

```python
catagories=['Q1','Q2','Q3','Q4']
product_a=[20,35,30,35]
product_b=[25,32,32,20]
product_c=[30,25,30,25]

x=np.arange(len(catagories))
width=0.25
plt.figure(figsize=(10,6))

plt.bar(x-width,product_a,width,label='Product A',color='#FF6B6B')
plt.bar(x,product_b,width,label='Product B',color='#4ECDC4')
plt.bar(x+width,product_c,width,label='Product C',color='#95E1D3')

plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("Quarterly Sales Comparison")
plt.xticks(x,catagories)
plt.legend()
plt.grid(True,alpha=0.3,axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(2, 1, figsize=(10,4))

ax[0].plot(x, np.sin(x))
ax[0].set_title("Sine Wave")

# Right plot
ax[1].plot(x, np.cos(x))
ax[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(2, 2, figsize=(8,6))

ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title("Sine")

ax[0, 1].plot(x, np.cos(x))
ax[0, 1].set_title("Cosine")

ax[1, 0].plot(x, np.tan(x))
ax[1, 0].set_title("Tangent")

ax[1, 1].plot(x, np.exp(x))
ax[1, 1].set_title("Exponential")

plt.tight_layout()
plt.show()
```

```python
regions = ["North", "South", "East", "West"]
products = ["Laptop", "Phone", "Tablet", "Monitor"]

data=pd.DataFrame({
    "Dates": pd.date_range("2026-01-01", periods=20),
    "Region": np.random.choice(regions, 20),
    "Products": np.random.choice(products,20),
    "Sales": np.random.randint(100,1000,20),
    "Profit": np.random.randint(20, 300, 20)
})

print(data)
```

```python
fig, ax = plt.subplots(1,4,figsize=(14,4))

ax[0].bar(data["Products"],data["Sales"],color='red',edgecolor='black')
ax[0].set_title("Product with Most Sales")
ax[0].set_xlabel("Prodcut")
ax[0].set_ylabel("Sales")

ax[1].bar(data["Products"],data["Profit"],color='blue',edgecolor='black')
ax[1].set_title("Product with Most Profit")
ax[1].set_xlabel("Prodcut")
ax[1].set_ylabel("Profit")

ax[2].plot(data["Dates"],data["Sales"])
ax[2].set_title("Sales over time")
ax[2].set_xlabel("Dates")
ax[2].set_ylabel("Sales")

ax[3].plot(data["Dates"],data["Profit"])
ax[2].set_title("Profit over time")
ax[2].set_xlabel("Dates")
ax[2].set_ylabel("Profit")
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
fig, axes=plt.subplots(2,1,figsize=(12,10))

data.plot(x="Dates",y="Sales",ax=axes[0],color='steelblue',linewidth=2)
axes[0].set_title("Sales overtime using Pandas",fontsize=12)
axes[0].set_xlabel("Dates")
axes[0].set_ylabel("Sales")
axes[0].grid(True,alpha=0.3)
axes[0].tick_params(axis='x',rotation=45)

region_sales=data.groupby("Region")["Sales"].sum()
region_sales.plot(kind='bar',ax=axes[1],color="coral",edgecolor="black")
axes[1].set_title("Sales by Region",fontsize=12)
axes[1].set_xlabel("Region")
axes[1].set_ylabel("Sales")
axes[1].grid(True,alpha=0.3,axis='y')

plt.tight_layout()
plt.show()
```
